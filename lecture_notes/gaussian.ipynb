```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n=1000
mean=np.array([0,0])
A=np.array([
    [1,0],
    [0,1]
])

X=np.random.multivariate_normal(
    mean=mean,
    cov=np.eye(2),
    size=n
)
from scipy.linalg import sqrtm
X=np.dot(X,sqrtm(A))

w,v=np.linalg.eig(A)
print('w=',w)
print('v=',v)
eig0=v[:,0]
eig1=v[:,1]

fig,ax=plt.subplots(figsize=(8,8))
ax.scatter(X[:,0],X[:,1])
ax.arrow(0,0,w[0]*eig0[0],w[0]*eig0[1])
ax.arrow(0,0,w[1]*eig1[0],w[1]*eig1[1])
plt.show()
```
